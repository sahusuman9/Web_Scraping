```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
with open("html/page1.html") as f:
    content = f.read()
```

```python
soup = BeautifulSoup(content,"html.parser")
```

```python
articles = soup.select("article.product_pod")
```

```python
items = []
for article in articles:
    title = article.find('h3').find('a')['title']
    price = article.select_one('p.price_color').text.split('Â£')[1]
    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    stock = ' '.join(article.select_one("p.instock").text.split())
    items.append([title,price,rating,stock])
```

```python
items
```

```python
df = pd.DataFrame(items, columns=['Book Title','Price','Rating','Stock Availability'])
```
